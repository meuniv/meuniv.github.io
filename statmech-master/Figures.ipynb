```python
from scipy.stats import binom, norm
from numpy import *
import matplotlib.pyplot as plot
from scipy.interpolate import interp1d
%matplotlib inline
```

```python
def createBinomialPlot(n):
    y = binom.pmf(list(range(n+1)),n,0.5)
    if (n < 100):
        f = interp1d(linspace(0,1,len(y)), y, 'cubic')
        y = f(linspace(0,1,100))
    y /= max(y)
    return y
```

```python
fig, axes = plot.subplots()
axes.plot(linspace(0,1,100), createBinomialPlot(10), color='black', ls=':')
axes.plot(linspace(0,1,101), createBinomialPlot(100), color='black', ls='--')
axes.plot(linspace(0,1,1001), createBinomialPlot(1000), color='black', ls='-')
axes.set_ylim([0, 1.1])
axes.legend(['N=10', 'N=100', 'N=1000'])
plot.show()
fig.savefig('binomial.svg')
```

```python
fig, axes = plot.subplots(nrows=2, ncols=2)
avalues = [0, 0.5, 0, 1]
bvalues = [0.5, 0.5, 2, 2]
labels = ['(a)', '(b)', '(c)', '(d)']
limit = 2
for ax, a, b, label in zip(axes.flatten(), avalues, bvalues, labels):
    ax.plot([-limit, limit], [-limit, limit], color='black', ls=':')
    x = linspace(-limit, limit, 100)
    ax.plot(x, tanh(a+b*x), color='black')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position(('data',0))
    ax.spines['left'].set_position(('data',0))
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    ax.text(-limit, 1, label)
    ax.set_xlim([-limit, limit])
    ax.set_ylim([-1.5, 1.5])
plot.show()
fig.savefig('mean_field.svg')
```
